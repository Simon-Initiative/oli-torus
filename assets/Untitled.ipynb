```python
import pandas as pd

# Load the CSV file
file_path = '~/Downloads/Book2.csv'
df = pd.read_csv(file_path)
print(df)

# Convert the DataFrame to a dictionary of records
records = df[['Sum of Actual Days', 'Year']].to_dict(orient='records')

# Display the first few records
records[:5]
```

```python
import altair as alt
# Create the box and whiskers plot
boxplot = alt.Chart(df).mark_boxplot(extent='min-max').encode(
    x=alt.X('Year:O', title='Year'),
    y=alt.Y('Sum of Actual Days:Q', title='Actual Days'),
    color=alt.Color('Year:O', legend=None)
)
expected_days_data = pd.DataFrame({
    'Year': [1, 2, 3, 4, 5],
    'Expected Days': [65, 135, 155, 155, 155]
})

line_chart = alt.Chart(expected_days_data).mark_line(color='red').encode(
    x=alt.X('Year:O', title='Year'),
    y=alt.Y('Expected Days:Q', title='Expected Days'),
)

# Combine the two charts
combined_chart = alt.layer(boxplot, line_chart).resolve_scale(
    
).properties(
    title='Box and Whiskers Plot of Actual Days with Expected Days Line',
    width=400
)

# Render the chart
combined_chart
```
